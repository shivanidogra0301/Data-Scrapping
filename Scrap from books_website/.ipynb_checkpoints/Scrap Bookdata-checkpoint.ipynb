```python
import requests
from bs4 import BeautifulSoup as soup

from urllib.parse import urljoin
```

```python
url="http://books.toscrape.com/"
book_infos=[]
```

```python
while True:
    
    r=requests.get(url)
    rsoup=soup(r.content,'html.parser')
    
    infos=rsoup.findAll('article',{"class":"product_pod"})
    for info in infos:
        book_img=info.find('img',{"class":"thumbnail"})
        book_price_response=info.find('p',{"class":"price_color"})
        book=[]
        book.append(book_img.attrs["src"]) #book url
        book.append(book_img.attrs["alt"]) #book title
        book.append(book_price_response.text) #book price
        book_infos.append(book)
    next_page=rsoup('li',{"class":"next"})
    
    print(next_page)
    if next_page is not None:
        next_page=next_page[0].a.attrs["href"]
        url=urljoin(url,next_page)
        
    else:
        break    
print(book_infos)    

```

```python
book_string=""
for idx,book_info in enumerate(book_infos):
    book_info_str=""
    for w in book_info:
#             print(w)
# to remoe unwanted commas between the words
            w=w.replace(',','')
            book_info_str+=w+','
    book_info_str=book_info_str[:-1]
    book_info_str+='\n'
    book_string+=book_info_str
print(book_string)
with open("books.csv",'w',encoding="utf-8") as f:
    f.write(book_string)
```
